```python
#analisando imagens
import cv2
from google.colab.patches import cv2_imshow

carregaFace =  cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml") #carregando algoritmos
carregaOlho =  cv2.CascadeClassifier("haarcascades/haarcascade_eye.xml")


imagem = cv2.imread("imagens/imagem5.jpg") #lendo a imagem
imagemCinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY) #transformando ela em cinza
faces = carregaFace.detectMultiScale(imagemCinza) #algoritmo lendo as faces na escala de cinza
print(faces) #mostrando as faces em escala de cinza

for(x, y, l, a) in faces: #condição de leitura das faces
  imagem = cv2.rectangle(imagem, (x, y), (x + l, y + a), (0, 200, 0), 4)
  localOlho = imagem[y:y + a, x:x + l]
  localOlhoCinza = cv2.cvtColor(localOlho, cv2.COLOR_BGR2GRAY)
  detectado = carregaOlho.detectMultiScale(localOlhoCinza)

for (ox, oy, ol, oa) in detectado:
  cv2.rectangle(localOlho, (ox, oy), (ox + ol, oy + oa), (0, 0, 255), 2)

cv2_imshow(imagem) #mostrar imagem analisada
cv2.waitKey()
```
