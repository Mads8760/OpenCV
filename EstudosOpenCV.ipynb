```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
imagem = cv2.imread("/content/troller01.jpg")
print(imagem)
```

```python
cv2_imshow(imagem)
print(imagem.shape)
```

```python
porcentagem = 10
largura = int(imagem.shape[1] - (imagem.shape[1] * porcentagem / 100))
altura = int(imagem.shape[0] - (imagem.shape[0] * porcentagem / 100))
dimensoes = (largura, altura)
imagem_red = cv2.resize(imagem,dimensoes,interpolation = cv2.INTER_AREA)
cv2_imshow(imagem_red)
print(imagem_red.shape)
```

```python
altura_img, largura_img = imagem_red.shape[:2]
print(altura_img,largura_img)
centro = (largura_img/2, altura_img/2 )
print(centro)
```

```python
angulo = 25
escala = 1.0
matriz_rotação = cv2.getRotationMatrix2D(centro, angulo,escala)
print(matriz_rotação)
```

```python
img_rot = cv2.warpAffine(imagem_red, matriz_rotação, (largura, altura))
cv2_imshow(img_rot)
```

```python
imagem_rot = cv2.rotate(imagem_red, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(imagem_rot)
```
