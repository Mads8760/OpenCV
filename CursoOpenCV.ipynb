```python
import cv2

img = cv2.imread("/content/Troller01.jpg")
img_suave = cv2.GaussianBlur(img, (5,5), 0)
cv2_imshow(img_suave)
```

```python
img_filter = cv2.bilateralFilter(img, 100, 100, 100)
cv2_imshow(img_filter)
```

```python
img_borda = cv2.Canny(imagem_png, 100, 200)
cv2_imshow(img_borda)
```

```python
img_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_suave = cv2.GaussianBlur(imagem, (5,5), 8)
cv2_imshow(img_suave)
```

```python
img_arestas = cv2.Canny(img_suave, 120, 410)
cv2_imshow(img_arestas)
```

```python
contorno, _ = cv2.findContours(img_arestas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for contornar in contorno:
  cv2.drawContours(imagem, [contornar], -1, (255,0,255),1)
cv2_imshow(imagem)
```
